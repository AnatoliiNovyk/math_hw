# Завдання 1.
В шаблоні наведені числові вектори $\overline{a}$ та $\overline{b}$. Порахуй наступне:
- сума $\overline{a}$ та $\overline{b}$
- різниця $\overline{a}$ та $\overline{b}$
- сума $\overline{a}$ та $\overline{b}^T$. Поясни отриманий результат.
- матричний добуток (dot product) $\overline{a}$ та $\overline{b}^T$.
- матричний добуток (dot product) $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- добуток Адамара (Hadamard product) $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- ділення $\overline{a}$ та $\overline{b}$. Поясни отриманий результат.
- ділення $\overline{a}$ та $\overline{b}^T$. Поясни отриманий результат.

````python
import numpy as np
# Код оголошення векторів не можна змінювати
a = np.array([[1, 2, 3, 4, 5]])
b = np.array([[1/2, 1, 2, 3, 4]])
````

````python
# 1. Сума a + b
sum_ab = a + b
print("1. Сума a + b:", sum_ab)
````

````python
# 2. Різниця a - b
diff_ab = a - b
print("2. Різниця a - b:", diff_ab)
````

````python
# 3. Сума a + b^T (транспонування не впливає на поелементні операції)
b_T = b.T  # Транспонування для одномірного масиву в NumPy не змінює вектор
sum_a_bT = a + b_T
print("3. Сума a + b^T:", sum_a_bT)
````

````python
# 4. Скалярний добуток a · b^T
dot_a_bT = np.dot(a, b_T)
print("4. Скалярний добуток a · b^T:", dot_a_bT)
````

````python
# 5. Скалярний добуток a · b
dot_ab = np.dot(a, b)
print("5. Скалярний добуток a · b:", dot_ab)
````

````python
# 6. Добуток Адамара a ⊙ b
hadamard_ab = a * b
print("6. Добуток Адамара a ⊙ b:", hadamard_ab)
````

````python
# 7. Ділення a ÷ b
div_ab = a / b
print("7. Ділення a ÷ b:", div_ab)
````

````python
# 8. Ділення a ÷ b^T
div_a_bT = a / b_T
print("8. Ділення a ÷ b^T:", div_a_bT)
````

````python
import numpy as np

# Визначимо вектор x
x = np.array([2, 1])

# 1. Зменшення в 2 рази по OX, збільшення в 3 рази по OY
M1 = np.array([[1/2, 0], [0, 3]])
M1x = np.dot(M1, x)
print("1. Зменшення по OX у 2 рази, збільшення по OY у 3 рази:")
print("M1 =", M1)
print("M1 * x =", M1x)

# 2. Відображення відносно початку координат
M2 = np.array([[-1, 0], [0, -1]])
M2x = np.dot(M2, x)
print("\n2. Відображення відносно початку координат:")
print("M2 =", M2)
print("M2 * x =", M2x)

# 3. Перенесення на -3 по OX, на 1 по OY (використовуємо гомогенні координати)
x_homogeneous = np.array([2, 1, 1])  # Гомогенні координати
M3 = np.array([[1, 0, -3], [0, 1, 1], [0, 0, 1]])
M3x = np.dot(M3, x_homogeneous)
M3x = M3x[:2]  # Беремо тільки x і y, відкидаємо 1
print("\n3. Перенесення на -3 по OX, на 1 по OY:")
print("M3 =", M3)
print("M3 * x =", M3x)

# 4. Зсув (shear) на 60° по OY
sqrt_3 = np.sqrt(3)
M4 = np.array([[1, 0], [sqrt_3, 1]])
M4x = np.dot(M4, x)
print("\n4. Зсув на 60° по OY:")
print("M4 =", M4)
print("M4 * x =", M4x)

# 5. Поворот на 30°
cos_30 = np.sqrt(3) / 2
sin_30 = 1 / 2
M5 = np.array([[cos_30, -sin_30], [sin_30, cos_30]])
M5x = np.dot(M5, x)
print("\n5. Поворот на 30°:")
print("M5 =", M5)
print("M5 * x =", M5x)

# 6. Об'єднання перетворень M5 * M4 * M2 * M1
M6 = np.dot(np.dot(np.dot(M5, M4), M2), M1)
M6x = np.dot(M6, x)
print("\n6. Об'єднане перетворення (M5 * M4 * M2 * M1):")
print("M6 =", M6)
print("M6 * x =", M6x)
````

# Завдання 2 (Альтернативне)

````python
import cv2 as cv
import urllib
import numpy as np
from google.colab.patches import cv2_imshow as cv_imshow

# Функція для завантаження зображення з URL (надана викладачем)
def read_image_by_url(url):
    req = urllib.request.urlopen(url)
    arr = np.asarray(bytearray(req.read()), dtype=np.uint8)
    img = cv.imdecode(arr, -1)
    return img

# URL зображення (публічне зображення котика з Вікіпедії)
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/4/4d/Cat_November_2010-1a.jpg/320px-Cat_November_2010-1a.jpg"

# Завантажуємо зображення з обробкою помилок
try:
    img = read_image_by_url(url)
    if img is None:
        raise ValueError("Не вдалося декодувати зображення.")
except Exception as e:
    print(f"Помилка завантаження зображення: {e}")
    print("Спробуй інший URL або завантаж зображення локально.")
    # Якщо завантаження не вдалося, можна завантажити зображення в Colab вручну:
    # from google.colab import files
    # uploaded = files.upload()
    # img = cv.imdecode(np.frombuffer(uploaded[list(uploaded.keys())[0]], np.uint8), -1)
    exit()
````

````python
# Показуємо оригінальне зображення
print("Оригінальне зображення:")
cv_imshow(img)

# Отримуємо розміри зображення
h, w = img.shape[:2]
````

````python
# Отримуємо розміри зображення
h, w = img.shape[:2]

# 1. Зменшення в 2 рази по OX, збільшення в 3 рази по OY
M1 = np.array([[1/2, 0, 0],
              [0, 3, 0],
              [0, 0, 1]], dtype=np.float32)
img1 = cv.warpPerspective(img, M1, (w, h))
print("\n1. Зменшення в 2 рази по OX, збільшення в 3 рази по OY:")
cv_imshow(img1)

# 2. Відображення відносно початку координат
M2 = np.array([[-1, 0, w],
               [0, -1, h],
               [0, 0, 1]], dtype=np.float32)
img2 = cv.warpPerspective(img, M2, (w, h))
print("\n2. Відображення відносно початку координат:")
cv_imshow(img2)

# 3. Перенесення на -3 по OX, на 1 по OY
M3 = np.array([[1, 0, -3],
               [0, 1, 1],
               [0, 0, 1]], dtype=np.float32)
img3 = cv.warpPerspective(img, M3, (w, h))
print("\n3. Перенесення на -3 по OX, на 1 по OY:")
cv_imshow(img3)

# 4. Зсув (shear) на 60° по OY
sqrt_3 = np.sqrt(3)
M4 = np.array([[1, 0, 0],
               [sqrt_3, 1, 0],
               [0, 0, 1]], dtype=np.float32)
img4 = cv.warpPerspective(img, M4, (w, h))
print("\n4. Зсув на 60° по OY:")
cv_imshow(img4)

# 5. Поворот на 30°
cos_30 = np.sqrt(3) / 2
sin_30 = 1 / 2
center_x, center_y = w / 2, h / 2
T1 = np.array([[1, 0, -center_x],
              [0, 1, -center_y],
              [0, 0, 1]], dtype=np.float32)
R = np.array([[cos_30, -sin_30, 0],
              [sin_30, cos_30, 0],
              [0, 0, 1]], dtype=np.float32)
T2 = np.array([[1, 0, center_x],
               [0, 1, center_y],
               [0, 0, 1]], dtype=np.float32)
M5 = T2 @ R @ T1
img5 = cv.warpPerspective(img, M5, (w, h))
print("\n5. Поворот на 30°:")
cv_imshow(img5)

# 6. Об'єднане перетворення (M5 * M4 * M2 * M1)
M6 = M5 @ M4 @ M2 @ M1
img6 = cv.warpPerspective(img, M6, (w, h))
print("\n6. Об'єднане перетворення (M5 * M4 * M2 * M1):")
cv_imshow(img6)
````
