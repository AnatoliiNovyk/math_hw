### ЗАВДАННЯ

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

# Завантаження даних
data = pd.read_csv('/content/drive/MyDrive/GOIT/dataset/weatherHistory.csv')
data = data.dropna()

# Вибір ознак і цільової змінної
features = ['Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Pressure (millibars)']
X = data[features]
y = data['Temperature (C)']

# Розділення даних
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Створення та навчання моделі
model = DecisionTreeRegressor(random_state=1)
model.fit(X_train, y_train)

# Прогноз для перших 10 рядків
first_ten_predictions = model.predict(X.head(10))
print("Прогнозовані температури для перших 10 рядків:")
print(first_ten_predictions)

# Фактичні значення
print("\nФактичні температури для перших 10 рядків:")
print(data['Temperature (C)'].head(10).values)
```
