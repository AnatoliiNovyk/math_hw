### ЗАВДАННЯ 1

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

# Завантаження даних
data = pd.read_csv('/content/drive/MyDrive/GOIT/dataset/weatherHistory.csv')
data = data.dropna()

# Вибір ознак і цільової змінної
features = ['Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)', 'Pressure (millibars)']
X = data[features]
y = data['Temperature (C)']

# Розділення даних
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Створення та навчання моделі
model = DecisionTreeRegressor(random_state=1)
model.fit(X_train, y_train)

# Прогноз для перших 10 рядків
first_ten_predictions = model.predict(X.head(10))
print("Прогнозовані температури для перших 10 рядків:")
print(first_ten_predictions)

# Фактичні значення
print("\nФактичні температури для перших 10 рядків:")
print(data['Temperature (C)'].head(10).values)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# Завантаження набору даних Iris
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

# Перетворення цільової змінної у назви видів
y = pd.Series(y).map({0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'})

# Перейменування стовпців
X.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Розділення даних на тренувальну і тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Створення моделі та навчання
model = DecisionTreeClassifier(random_state=1)
model.fit(X_train, y_train)

# Дані для прогнозу
data_to_predict = pd.DataFrame({
    'SepalLengthCm': [5.9, 6.9, 5.0, 5.2, 5.8],
    'SepalWidthCm': [3.2, 3.2, 3.4, 3.5, 2.7],
    'PetalLengthCm': [4.8, 5.7, 1.6, 1.5, 4.1],
    'PetalWidthCm': [1.8, 2.3, 0.4, 0.2, 1.0]
})

# Прогноз
predictions = model.predict(data_to_predict)
print("Прогнозовані види для рядків з Id 71, 121, 27, 28, 68:")
print(predictions)

# Фактичні значення
actual = ['Iris-versicolor', 'Iris-virginica', 'Iris-setosa', 'Iris-setosa', 'Iris-versicolor']
print("\nФактичні види для рядків з Id 71, 121, 27, 28, 68:")
print(actual)
```
